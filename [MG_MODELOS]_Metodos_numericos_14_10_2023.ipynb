###Matriz identidad


```r
N<-matrix(0,3,3)
N[row(N)==col(N)]<-1
print(N)
```

```r
is.matrix(N)
```

```r
## que pasa que quisiese que la diagonal superior sea uno

N<-matrix(0,3,3)
N[row(N)<=col(N)]<-1
print(N)
```

```r
## que pasa que quisiese que la diagonal superior sea uno

N<-matrix(0,3,3)
N[row(N)>=col(N)]<-1
print(N)
```

###Suma de matrices

```r
M<-matrix(1:6,byrow=TRUE,nrow=3,ncol=2)
N<-matrix(7:12,nrow=3,ncol=2)
print(M)
print(N)
```

```r
MS<-M+N
print(MS)
```

```r
MS<-M-N
print(MS)
```

###Multiplicacion de matrices

```r
# multiplicacion entre matrices
M<-matrix(1:6,byrow=TRUE,nrow=3,ncol=2)
N<-matrix(7:10,nrow=2,ncol=2)
MU<- M%*%N
print(MU)
```

```r
# Multiplicacion escalar
ME<- 2*M
print(ME)
```

```r
M=matrix(c(1,1,0,4,2,4,3,3,1),nrow=3,ncol=3,byrow=TRUE)
M
```

```r
medias=matrix(0,ncol=3,nrow=3)
medias[,1]<-mean(M[,1])
medias[,2]<-mean(M[,2])
medias[,3]<-mean(M[,3])
medias
```

```r
dif=M-medias
dif
```

```r
fact<-(medias%*%t(medias))
fact
```

```r
n<-3
((dif%*%t(dif))/(n-1))-fact
```
